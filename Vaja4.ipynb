```python

# verzija 1
from qiskit import QuantumCircuit
from qiskit.primitives import StatevectorSampler
from math import pi
qc = QuantumCircuit(8)
qc.rx(pi/2,range(8))
qc.rx(-pi/3,0) #1
qc.rx(pi/3,1) #0
qc.rx(pi/3,2) #0
qc.rx(pi/3,3) #0
qc.rx(pi/3,4) #0
qc.rx(pi/3,5) #0
qc.rx(-pi/3,6) #1
qc.rx(pi/3,7) #0

#01000001

qc.measure_all()

sampler = StatevectorSampler()
result = sampler.run([qc], shots=1000000).result()
#result = sampler.run([qc], shots=10).result()
print(result[0].data.meas.get_counts())
```

```python
from qiskit.visualization import plot_histogram

counts = result[0].data.meas.get_counts()
decimal_counts = dict(
    sorted(
        ((str(int(k, 2)), v) for k, v in counts.items()),
        key=lambda item: int(item[0])
    )
)
plot_histogram(decimal_counts)
```
